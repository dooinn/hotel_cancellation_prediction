```python
import requests
from bs4 import BeautifulSoup
```

```python
r = requests.get('https://www.booking.com/hotel/pt/lisbon-city.html?lang=en-us&soz=1&lang_changed=1#tab-reviews').content
soup = BeautifulSoup(r, 'html.parser')
print(soup.prettify())
```

```python
# Find the div element with id 'review_list_page_container'
review_list_div = soup.find('div', id='review_list_page_container')

# Now review_list_div contains the desired div block along with all of its content.
# You can print it or process it further.
print(review_list_div)
```
